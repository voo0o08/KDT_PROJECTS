```python
import pandas as pd
import numpy as np
import csv
```

## 국민연금 가입

```python
f1 = "국민연금공단_국민연금 가입 사업장 내역_20240119.csv"
pensionDF = pd.read_csv(f1, encoding="cp949")
```

```python
pensionDF.head()
```

```python
pensionDF.columns
```

사용할 데이터 col을 뽑아보자

```python
pensionDF = pensionDF[["사업장명", "사업장지번상세주소", "가입자수", "당월고지금액"]]
```

"사업장지번상세주소" -> "본사주소"로 변경

```python
pensionDF.columns = ["사업장명", "본사주소", "가입자수", "당월고지금액"]
pensionDF.columns
```

### 주소분리

```python
pensionDF.sort_values("당월고지금액")
```

```python
addr = pd.DataFrame(pensionDF["본사주소"])
addr = addr["본사주소"].apply(lambda x : x.split()[:2])
addr =	pd.DataFrame(addr.tolist(), columns=('시도', '군구'))
addr.head()
```

```python
pensionDF[["시도", "군구", "나머지 주소"]] = pensionDF["본사주소"].str.split(' ', n=2, expand=True)
pensionDF.head()
```

```python
pensionDF = pensionDF.drop("본사주소", axis = True)
```

```python
pensionDF["시도"].unique()
```

## <hr>SW 기업 DF뽑기

```python
f2 = "정보통신산업진흥원_SW산업정보종합시스템 사업자정보DB_20221031.csv"
swDF = pd.read_csv(f2, encoding="cp949")
```

```python
swDF
```

```python
swDF.columns
```

```python
swDF = swDF[["회사명", "본사주소"]]
swDF.head()
```

회사명 -> 사업장명으로 변경

```python
swDF.columns = ["사업장명", "본사주소"]
swDF.head()
# f.reindex(columns=['http_status', 'user_agent'])
```

```python
swDF[["시도", "군구", "나머지 주소"]] = pensionDF["본사주소"].str.split(' ', n=2, expand=True)
swDF.head()
```

```python
swDF = swDF.drop("본사주소", axis=True)
```

```python
swDF["시도"].unique()
```

## join 사업장명을 기준으로
-swDF / pensionDF

```python
mergeDF = pd.merge(swDF, pensionDF, on=["사업장명", "시도"])
mergeDF.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib
```

```python
Data = pd.read_csv('전_산업의_인력_채용_시_우대_예정__자격증_20240129190428.csv',  header=[0, 1, 2])
index=Data['자격증별(1)']
index
```

```python
Data = Data[['2020', '2021', '2022']]
```

```python
Data2020 = Data['2020']
Data2020 =Data2020['데이터산업']
Data2020 = pd.concat([index, Data2020], axis = 1)
Data2020.columns = ['자격증별', '응답수 (개)', '비중(%)']
Data2020=Data2020.set_index('자격증별')
```

```python
# 결측치 확인
Data2020 = Data2020.replace('-', '0')
Data2020['비중(%)'] = Data2020['비중(%)'].astype(float)
Data2020['응답수 (개)'] = Data2020['응답수 (개)'].astype(int)
Data2020['비중(%)']
Data2020_without_total = Data2020.drop(labels = '사례수')
Data2020_without_total
```

```python
color_set=['#FCD3DC', '#F57793', '#F7E7C3', '#AED99E', '#96D4D0', '#629DD1', '#6264D1', '#B062D1', '#F7C3E6', '#F7C3C3', '#97A4DE', '#B4BBD9']
plt.figure(figsize=(20,10))
plt.pie(Data2020_without_total['비중(%)'], colors = color_set)
plt.legend(Data2020_without_total.index)
```

```python
color_set=['#FCD3DC', '#F57793', '#F7E7C3', '#AED99E', '#96D4D0', '#629DD1', '#6264D1', '#B062D1', '#F7C3E6', '#F7C3C3', '#97A4DE', '#B4BBD9']
plt.figure(figsize=(20,10))
plt.title('[2020년 산업의 인력 채용 시 우대 (예정) 자격증]', fontsize ='xx-large')
plt.ylabel('자격증별', fontsize = 'xx-large')
Data2020_without_total['응답수 (개)'].plot(kind='barh', color = color_set, fontsize = 'xx-large')
```

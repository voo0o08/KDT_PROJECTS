```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
DF = pd.read_csv('데이터산업의_필요_인력_채용_시_우대_기술_20240129174057.csv', header=[0, 1, 2], skiprows=0)
index = DF['우대사항별(1)']
index
```

```python
year2020 = DF["2020"]
year2020 = year2020['데이터산업 전체']
year2020 = pd.concat([index, year2020], axis = 1)
year2020.columns = ['우대사항별', '응답기업수 (개)', '비중(%)']
year2020
year2020.set_index('우대사항별')
year2020 = year2020.set_index('우대사항별')
year2020
```

```python
year2020 = year2020.sort_values(by='응답기업수 (개)', ascending=False)
```

```python
year2020_without_total = year2020.drop(labels = '전체')
year2020_without_total['비중(%)'] = year2020_without_total['비중(%)'].astype(float)
year2020_without_total['응답기업수 (개)']=year2020_without_total['응답기업수 (개)'].astype(int)
```

```python
color_set=['#FCD3DC', '#F57793', '#F7E7C3', '#AED99E', '#96D4D0', '#629DD1', '#6264D1', '#B062D1', '#F7C3E6', '#F7C3C3', '#97A4DE', '#B4BBD9']
plt.figure(figsize=(20,10))
plt.pie(year2020_without_total['비중(%)'], colors = color_set)
plt.title('[2020년 데이터 산업의 필요 인력 채용 시 우대 기술]')
plt.legend(year2020_without_total.index)
```

```python
# 상위 다섯개 항목만...
year2020_top5 = year2020_without_total[:5]
index=year2020_top5.index
```

```python
color_set1=['#F57793', '#FCD3DC', '#F7E7C3', '#AED99E', '#96D4D0']
plt.figure(figsize=(20,10))
plt.ylabel('우대사항별', fontsize = 'xx-large')
plt.title('[2020년 데이터 산업의 필요 인력 채용 시 우대 기술 Top 5]', fontsize = 'xx-large')
year2020_top5['응답기업수 (개)'].plot(kind='barh', color = color_set1, fontsize = 'xx-large')
plt.legend(['분석 프로그래밍', '통계툴 분석 능력', 'SQL 프로그래밍', '데이터 시각화', '텍스트·자연어 분석'])
plt.show()
```

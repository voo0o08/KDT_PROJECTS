```python
workforce = pd.read_csv('데이터산업의_데이터직무별_인력_현황_20240129171851.csv', usecols=['데이터직무별(1)','2018', '2019', '2020','2021','2022'], index_col='데이터직무별(1)')
workforce.dtypes
```

```python
workforce=workforce.drop(labels=['데이터직무별(1)'])
```

```python
data_analyst_total = workforce.loc['전체']
data_analyst = workforce.loc['데이터 분석가']
data_analyst_total=data_analyst_total.astype(int)
```

```python
data_analyst=data_analyst.astype(int)
```

```python
x_year = [2018, 2019, 2020, 2021, 2022]
color_set = ['#DCE8F7','#C7E0ED', '#A8C5E0', '#83B2DE', '#62A5E0']
plt.figure(figsize=(5,6))
plt.bar(x_year, data_analyst_total, color = color_set)
plt.title('[2018~2022년 데이터 직무별 인력 추이]')
plt.show()
```

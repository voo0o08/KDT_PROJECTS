```python
pd.read_csv('향후_데이터산업의_데이터직무_필요_인력_20240129172051.csv')
data_need = pd.read_csv('향후_데이터산업의_데이터직무_필요_인력_20240129172051.csv', usecols=['데이터직무별(1)','2018', '2019', '2020','2021','2022'], index_col='데이터직무별(1)')
data_need_5years = pd.read_csv('향후_데이터산업의_데이터직무_필요_인력_20240129172051.csv', usecols=['데이터직무별(1)','2018.1', '2019.1', '2020.1','2021.1','2022.1'], index_col='데이터직무별(1)')
```

```python
#차년도 필요 인력 (명)
data_need.dtypes
```

```python
# 향후 5년 내 필요 인력 (명)
data_need_5years
```

```python
data_need=data_need.drop(labels=['데이터직무별(1)'])
data_need
```

```python
data_need_5years=data_need_5years.drop(labels=['데이터직무별(1)'])
data_need_5years.columns = [2018, 2019, 2020, 2021, 2022]
data_need_5years
```

```python
#차년도 필요 인력 (명) - 전체 / 데이터 분석가
data_total = data_need.loc['전체'].astype(int)
data_need_analyst = data_need.loc['데이터 분석가'].astype(int)
```

```python
data_total
```

```python
data_need_analyst
```

```python
plt.subplot(1,2,1)
x_year = [2018, 2019, 2020, 2021, 2022]
plt.bar(x_year, data_total, color = ['#C7E0ED','#C7E0ED', '#C7E0ED', '#62A5E0', '#C7E0ED'])
plt.xticks(x_year)
plt.title('[2018~2022년 차년도 필요 인력 (명) - 전체]')
plt.show()
```

```python
plt.subplot(1,2,2)
x_year = [2018, 2019, 2020, 2021, 2022]
plt.bar(x_year, data_need_analyst, color = ['#FBD2DB', '#FBD2DB', '#FBD2DB', '#FBD2DB', '#F76884'])
plt.title('[2018~2022년 차년도 필요 인력 (명) - 데이터 분석가]')
plt.xticks(x_year)
plt.show()
```

```python
# 향후 5년 내 필요 인력 (명) -> 전체/ 데이터 분석가
data_total_5years = data_need_5years.loc['전체']
data_need_5years_analyst = data_need_5years.loc['데이터 분석가']
```
